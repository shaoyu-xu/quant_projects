# Base Functions of Series and DataFrame

## 1.Rebuild Index

```python
import pandas as pd
import numpy as np
```

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c']) # create a pandas Series with specified index
```

```python
obj
```

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e']) # reindex the Series to a new index, introducing NaN for missing values
```

```python
obj2
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4]) # create a new Series with non-consecutive indices
```

```python
obj3
```

```python
obj3.reindex(np.arange(6), method='ffill') # reindex the Series to a new index range, using forword fill methon to fill missing values
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3,3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California']) # create a DataFrame with specified index and columns
```

```python
frame
```

```python
frame2 = frame.reindex(index=['a', 'b', 'c', 'd']) # reindex the DataFrame to a new index, introducing NaN for missing values
```

```python
frame2
```

```python
states = ['Texas', 'Utah', 'California'] # create a list of states
```

```python
frame.reindex(columns=states) # reindex the DataFrame to a new column order, introducing NaN for missing values
```

```python
frame.reindex(states, axis='columns') # reindex the DataFrame to a new column order using axis parameter, introducing NaN for missing values
```

```python
frame.loc[['a', 'd', 'c'], ['California', 'Texas']] # select specific rows and columns from the DataFrame using .loc
```

## Drop items on the specified axis

```python
obj = pd.Series(np.arange(5.), index=['a', 'b', 'c', 'd', 'e']) # create a new Series with specified index
```

```python
obj
```

```python
new_obj = obj.drop('c') # drop the item with index 'c' from the Series
```

```python
new_obj
```

```python
obj.drop(['d', 'c']) # drop multiple items from the Series by specifying a list of indices
```

```python
data = pd.DataFrame(np.arange(16).reshape((4,4)),
                    index=['Ohio', 'Colorado', 'Utah', 'New york'],
                    columns=['one', 'two', 'three', 'four']) # create a DataFrame
```

```python
data
```

```python
data.drop(index=['Colorado', 'Ohio']) # drop the rows with specified indices from the DataFrame
```

```python
data.drop(columns=['two']) # drop the column with specified name from the DataFrame
```

```python
data.drop('two', axis=1) # drop the column with specified name from the DataFrame using axis parameter
```

```python
data.drop(['two', 'four'], axis='columns') # drop multiple columns from the DataFrame by specifying a list of column names and using axis parameter
```

## Indexing, selection, and filtering

```python
obj = pd.Series(np.arange(4.), index=['a', 'b', 'c', 'd']) # create a new Series with specified index
```

```python
obj
```

```python
obj['b'] # select the value at index 'b' from the Series
```

```python
obj.iloc[1] # select the value at position 1 from the Series using positional indexing
```

```python
obj[2:4] # select a slice of the Series from position 2 to 4 (exclusive)
```

```python
obj[['b', 'a', 'd']] # select multiple values from the Series by specifying a list of indices
```

```python
obj[[1,3]] # select multiple values from the Series by specifying a list of positions
```

```python
obj[obj<2] # select values from the Series that are less than 2 using boolean indexing
```

```python
obj.loc[['b', 'a', 'd']] # select multiple values from the Series using .loc with a list of indices
```

```python
obj1 = pd.Series([1, 2, 3], index=[2, 0, 1]) # create a new Series with specified index
```

```python
obj2 = pd.Series([1, 2, 3], index=['a', 'b', 'c']) # create a new Series with specified index
```

```python
obj1
```

```python
obj2
```

```python
obj1[[0, 1, 2]] # select multiple values from the Series by specifying a list of positions
```

```python
obj2[[0, 1, 2]] # select multiple values from the Series by specifying a list of positions
```

```python
obj2.loc[[0,1]]
```

```python
obj1.iloc[[0, 1, 2]] # select multiple values from the Series by specifying a list of positions using .iloc
```

```python
obj2.iloc[[0, 1, 2]] # select multiple values from the Series by specifying a list of positions using .iloc
```

```python
obj2.loc['b':'c'] # select a slice of the Series from index 'b' to 'c' (inclusive) using .loc
```

```python
data = pd.DataFrame(np.arange(16).reshape((4,4)),
                    index=['Ohio', 'Colorado', 'Utah', 'New York'],
                    columns=['one', 'two', 'three', 'four']) # create a DataFrame
```

```python
data
```

```python
data['two'] # select the 'two' column from the DataFrame
```

```python
data[['three', 'one']] # select multiple columns from the DataFrame by specifying a list of column names
```

```python
data[:2] # select the first two rows of the DataFrame
```

```python
data[data['three']>5] # select rows from the DataFrame where the 'three' column is greater than 5
```

```python
data < 5 # create a DataFrame of boolean values indicating whether each element in the original DataFrame is less than 5
```

```python
data[data < 5] = 0 # set all values in the DataFrame that are less than 5 to 0
```

```python
data
```

```python
data
```

```python
data.loc['Colorado'] # select the row with index 'Colorado' from the DataFrame using .loc
```

```python
data.loc[['Colorado', 'New York']] # select multiple rows from the DataFrame by specifying a list of indices using .loc
```

```python
data.loc['Colorado', ['two', 'three']] # select specific rows and columns from the DataFrame using .loc
```

```python
data.iloc[2] # select the row at position 2 from the DataFrame using .iloc
```

```python
data.iloc[2,[3, 0, 1]] # select specific rows and columns from the DataFrame using .iloc
```

```python
data.iloc[[1, 2], [3, 0, 1]] # select specific rows and columns from the DataFrame using .iloc with lists of positions
```

```python
data.loc[:'Utah', 'two'] # select rows up to 'Utah' and the 'two' column from the DataFrame using .loc
```

```python
data.iloc[:,:3][data.three > 5] # select all rows and the first three columns from the DataFrame where the 'three' column is greater than 5 using .iloc and boolean indexing
```

```python
data.loc[data.three >= 2] # select all rows from the DataFrame where the 'three' column is greater than or equal to 2 using .loc and boolean indexing
```

```python
data.loc[:,'one'] = 1 # set all values in the 'one' column to 1 using .loc
```

```python
data
```

```python
data.iloc[2] = 5 # select the row at position 2 from the DataFrame using .iloc
```

```python
data
```

```python
data.loc[data['four'] > 5] = 3 # set all rows in the DataFrame where the 'four' column is greater than 5 to 3 using .loc and boolean indexing
```

```python
data
```

```python
data.loc[data.three == 5,'three'] = 6 # set all values in the 'three' column to 6 where the 'three' column is equal to 5 using .loc and boolean indexing
```
