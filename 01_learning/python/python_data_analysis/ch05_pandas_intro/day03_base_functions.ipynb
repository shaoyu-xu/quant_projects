# Base Functions of Series and DataFrame

## 1.Rebuild Index

```python
import pandas as pd
import numpy as np
```

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c']) # create a pandas Series with specified index
```

```python
obj
```

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e']) # reindex the Series to a new index, introducing NaN for missing values
```

```python
obj2
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4]) # create a new Series with non-consecutive indices
```

```python
obj3
```

```python
obj3.reindex(np.arange(6), method='ffill') # reindex the Series to a new index range, using forword fill methon to fill missing values
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3,3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California']) # create a DataFrame with specified index and columns
```

```python
frame
```

```python
frame2 = frame.reindex(index=['a', 'b', 'c', 'd']) # reindex the DataFrame to a new index, introducing NaN for missing values
```

```python
frame2
```

```python
states = ['Texas', 'Utah', 'California'] # create a list of states
```

```python
frame.reindex(columns=states) # reindex the DataFrame to a new column order, introducing NaN for missing values
```

```python
frame.reindex(states, axis='columns') # reindex the DataFrame to a new column order using axis parameter, introducing NaN for missing values
```

```python
frame.loc[['a', 'd', 'c'], ['California', 'Texas']] # select specific rows and columns from the DataFrame using .loc
```
