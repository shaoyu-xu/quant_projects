## 5.function application and mapping

```python
frame = pd.DataFrame(np.random.standard_normal((4,3)),
                     columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon']) # create a DataFrame with random values
```

```python
frame
```

```python
np.abs(frame) 
```

```python
def f1(x):
    return x.max() - x.min() # define a function to calculate the range of values in a Series
```

```python
frame.apply(f1) # apply the function to each column of the DataFrame
```

```python
frame.apply(f1, axis='columns') # apply the function to each row of the DataFrame
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max']) # define a function to calculate the minimum and maximum of values in a Series
```

```python
frame.apply(f2) # apply the function to each column of the DataFrame
```

```python
def my_format(x):
    return f'{x:.2f}' # define a function to format a float to two decimal places
```

```python
frame.map(lambda x: f'{x:.2f}') # apply the function to each element of the DataFrame
```

```python
frame['e'].map(my_format) # apply the function to the 'e' column of the DataFrame
```

## 6.Sorting and Ranking

```python
obj = pd.Series(np.arange(4), index=['d', 'a', 'b', 'c']) # create a Series
```

```python
obj
```

```python
obj.sort_index() # sort the Series by index
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2,4)),
                     index=['three', 'one'],
                     columns=['d', 'a', 'b', 'c']) # create a DataFrame
```

```python
frame
```

```python
frame.sort_index() # sort the DataFrame by index
```

```python
frame.sort_index(axis='columns') # sort the DataFrame by columns
```

```python
frame.sort_index(axis='columns', ascending=False) # sort the DataFrame by columns in descending order
```

```python
obj = pd.Series([4, 7, -3, 2]) # create a Series
```

```python
obj.sort_values() # sort the Series by values
```

```python
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2]) # create a Series with NaN values
```

```python
obj.sort_values() # sort the Series by values, ignoring NaN values
```

```python
obj.sort_values(na_position='first') # sort the Series by values, placing NaN values at the beginning
```

```python
frame = pd.DataFrame({'b': [4, 7, -3, 2], 'a': [0, 1, 0, 1]}) # create a DataFrame
```

```python
frame
```

```python
frame.sort_values('b') # sort the DataFrame by column 'b'
```

```python
frame.sort_values(['a', 'b']) # sort the DataFrame by columns 'a' and 'b'
```

```python
obj = pd.Series([7, -5, 7, 4, 2, 0, 4]) # create a Series
```

```python
obj.rank() # rank the Series by values
```

```python
obj.rank(method='first') # rank the Series by values, placing ties first
```

```python
obj.rank(ascending=False) # rank the Series by values in descending order
```

```python
frame = pd.DataFrame({'b': [4.3, 7, -3, 2], 'a': [0, 1, 0, 1],
                      'c': [-2, 5, 8, -2.5]}) # create a DataFrame
```

```python
frame
```

```python
frame.rank(axis='columns') # rank the DataFrame by columns
```
