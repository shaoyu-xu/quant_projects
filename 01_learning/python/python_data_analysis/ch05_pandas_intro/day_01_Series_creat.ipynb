```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000} # create a dictionary
obj3 = pd.Series(sdata) # create a Series from the dictionary
obj3 
```

```python
obj3.to_dict() # convert the Series to a dictionary
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas'] # create a list of states
obj4 = pd.Series(sdata, index=states) # create a Series with specified index
obj4
```

```python
pd.isnull(obj4) # check for null values
```

```python
pd.notna(obj4) # check for non-null values
```

```python
obj4.isna() # check for null values using Series method
```
