```python
import pandas as pd
import numpy as np
```

```python
data = {'state':['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Neveda', 'Neveda'],
        'year':[2000, 2001, 2002, 2001, 2002, 2003],
        'pop':[1.5, 1.7, 3.6, 2.4, 2.9, 3.2]} # dictionary
frame = pd.DataFrame(data) # create a dataframe from the dictionary
frame
```

```python
frame.head() # display the first 5 rows of the dataframe
```

```python
frame.tail() # display the last 5 rows of the dataframe
```

## Index Objects

```python
obj = pd.Series(np.arange(3), index=['a', 'b', 'c']) # create a pandas Series with default integer index
```

```python
index = obj.index # create a new index object
```

```python
index
```

```python
index[1:] # slice the index object
```

```python
index[1] = 'd'  # This will raise an error because Index objects are immutable
```

```python
labels = pd.Index(np.arange(3)) # create a new index object with a range of integers
```

```python
labels
```

```python
obj2 = pd.Series([1.5, -2.5, 0], index=labels) # create a pandas Series with a specified index
```
