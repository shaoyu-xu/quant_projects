```python
# 创建模拟数据
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'math': [85, 92, 78, 88],
    'english': [88, 79, 94, 82],
    'python': [90, 85, 80, 95]
})

df['total'] = df['math'] + df['english'] + df['python']
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='bar', x='name', y=['math', 'english', 'python'])
plt.title('Student Scores by Subject')
plt.ylabel('Score')
plt.show()
```
